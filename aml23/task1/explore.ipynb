```python
import pandas as pd
from pandas_profiling import ProfileReport
```

```python
x_train = pd.read_csv("data/x_train.csv").drop(["id"], axis=1)
x_train
```

```python
profile = ProfileReport(x_train, title="Data Exploration Task 01", minimal=True) # Disable heavy computations
profile.config.correlations["pearson"].calculate = True
```

```python
profile.to_file("explore.html")
```
